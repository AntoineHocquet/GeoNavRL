```python
# üìç OSM Graph Preview for GeoNavRL
# =================================

# üì¶ Imports (robust for VS Code / Ubuntu setups)
import osmnx as ox
import matplotlib.pyplot as plt
import os

# Optional: force interactive backend (uncomment only if needed)
# plt.switch_backend('TkAgg')  # or 'Qt5Agg' (Linux GUI)
# import matplotlib
# print(matplotlib.get_backend())  # debug the backend
```

```python
# üìå Settings
place_name = "Gesundbrunnen, Berlin"
network_type = "walk"
save_plot = True

# üåç Download the OSM walking graph
print(f"‚è≥ Downloading OSM graph for: {place_name}")
G = ox.graph_from_place(place_name, network_type=network_type)
print("‚úÖ Graph downloaded")
```

```python
# üñºÔ∏è Preview the graph
fig, ax = ox.plot_graph(G, bgcolor="white", node_color="red", edge_color="gray",
                        node_size=10, edge_linewidth=0.5, show=True, close=False)

# üíæ Save figure
if save_plot:
    output_dir = "../outputs"
    os.makedirs(output_dir, exist_ok=True)
    output_path = f"{output_dir}/osm_graph_{place_name.replace(' ', '_').replace(',', '')}.png"
    fig.savefig(output_path, dpi=300, bbox_inches="tight")
    print(f"üñºÔ∏è Graph saved to: {output_path}")
```
